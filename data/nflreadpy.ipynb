```python
import nflreadpy


draft_data = nflreadpy.load_draft_picks(seasons = range(2014,2026))[[
    'pfr_player_id','cfb_player_id','season' ,'round', 'pick', 'team', 'pfr_player_name', 'position', 'category', 'college', 'w_av', 'car_av', 'dr_av'
    ]].to_pandas()

draft_data

combine_data = nflreadpy.load_combine(seasons=range(2014, 2026)).to_pandas()
combine_data = combine_data[combine_data.pos.isin(['WR', 'QB', 'RB', 'TE'])]#dropna(subset = 'draft_round')
#[combine_data.pos.isin(['WR', 'QB', 'RB', 'TE'])]
combine_data.reset_index(inplace=True, drop = True)
```

```python
combine_data
```

```python
combine_data[(combine_data.season== 2025) & (combine_data.pos == 'WR')]
```

```python
draft_data_skill = draft_data[draft_data.position.isin(['WR', 'RB', 'QB', 'TE'])]
draft_data_skill
```

```python
import pandas as pd
merged_data = pd.merge(
    draft_data_skill,
    combine_data,
    left_on=['cfb_player_id', 'pfr_player_name','season'],
    right_on=['cfb_id','player_name','season'],
    how='left'
)

merged_data = merged_data[['pfr_player_id', 'cfb_player_id', 'season', 'round', 'pick', 'team', 'pfr_player_name', 'position', 'category', 'college',
            'w_av', 'car_av', 'dr_av', 'ht', 'wt', 'forty', 'bench', 'vertical', 'broad_jump', 'cone', 'shuttle']]

merged_data
```

```python
player_data = nflreadpy.load_players()['pfr_id', 'birth_date', 'height', 'weight', 'rookie_season', 'last_season']

merged_data_player = pd.merge(
    merged_data,
    player_data.to_pandas(),
    left_on='pfr_player_id',
    right_on='pfr_id',
    how='left'
)

merged_data_player.birth_date = pd.to_datetime(merged_data_player.birth_date)
merged_data_player
```

```python
merged_data_player['birth_year'] = merged_data_player['birth_date'].dt.year
display(merged_data_player[['pfr_player_name', 'birth_date', 'birth_year']].head())
```

```python
merged_data_player['age'] = merged_data_player['season'] - merged_data_player['birth_year']
merged_data_player
```
