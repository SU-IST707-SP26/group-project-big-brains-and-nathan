```python
import pandas as pd
combine = pd.read_csv("draft_combine_nflreadpy.csv").iloc[:,1:]
combine.head()
```

```python
draft = pd.read_csv("draft_picks_df_cfbd.csv")
draft['position'] = draft['position'].replace('Quarterback','QB')
draft['position'] = draft['position'].replace('Running Back','RB')
draft['position'] = draft['position'].replace('Wide Receiver','WR')
draft['position'] = draft['position'].replace('Tight End','TE')
positions = ['QB','RB','WR','TE']
draft = draft[draft['position'].isin(positions)]
draft.head()
```

```python
pd.set_option('display.max_columns', None)
```

```python
stats = pd.read_csv("stats_merged_df.csv")
stats = stats[stats['position'].isin(positions)]
stats.head()
```

```python
stats['college_season_played'] = stats.groupby('player_id')['season'].transform(lambda x: x - x.min() + 1)
stats.sort_values(['player_id', 'season'], ascending=True)
```

```python
stats.columns
```

```python
import numpy as np

stats1 = stats.copy()

stats1 = stats1.sort_values(['player_id', 'season'], ascending=True)

# Taking the Weighted Mean of All Statistics in a player's college career
# Weights are determined by their college season (ie Stats from
# first season played are weighted as 1 and 4th season
# played are weighted as 4, etc.)
stats1 = stats1.groupby(['player_id', 'name', 'position']).agg(
    team = ('team', 'last'),
    season = ('season', 'max'),
    conference = ('conference', 'last'),
    usage_passing_downs = ('usage_passing_downs',
                  lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    usage_standard_downs = ('usage_standard_downs',
                  lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    usage_third_down = ('usage_third_down',
                  lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    usage_second_down = ('usage_second_down',
                  lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    usage_first_down = ('usage_first_down',
                  lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    usage_rush = ('usage_rush',
                  lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    usage_var_pass = ('usage_var_pass',
                  lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    usage_overall = ('usage_overall',
                  lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    avg_ppa_passing_downs = ('avg_ppa_passing_downs',
                  lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    avg_ppa_standard_downs = ('avg_ppa_standard_downs',
                  lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    avg_ppa_third_down = ('avg_ppa_third_down', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    avg_ppa_second_down = ('avg_ppa_second_down', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    avg_ppa_first_down = ('avg_ppa_first_down', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    avg_ppa_rush = ('avg_ppa_rush', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    avg_ppa_var_pass = ('avg_ppa_var_pass', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    avg_ppa_all = ('avg_ppa_all', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    total_ppa_passing_downs = ('total_ppa_passing_downs',
                  lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    total_ppa_standard_downs = ('total_ppa_standard_downs', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    total_ppa_third_down = ('total_ppa_third_down', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    total_ppa_second_down = ('total_ppa_second_down', lambda x: np.average(x, weights = stats1.loc[x.index, 'college_season_played'].fillna(0))),
    total_ppa_first_down = ('total_ppa_first_down', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    total_ppa_rush = ('total_ppa_rush', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    total_ppa_var_pass = ('total_ppa_var_pass', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    total_ppa_all = ('total_ppa_all', lambda x: np.average(x,weights = stats1.loc[x.index, 'college_season_played'].fillna(0))),
    fumbles_FUM = ('fumbles_FUM', lambda x: np.average(x, weights = stats1.loc[x.index, 'college_season_played'].fillna(0))),
    fumbles_LOST = ('fumbles_LOST', lambda x: np.average(x, weights = stats1.loc[x.index, 'college_season_played'].fillna(0))),
    fumbles_REC = ('fumbles_REC', lambda x: np.average(x, weights = stats1.loc[x.index, 'college_season_played'].fillna(0))),
    passing_ATT = ('passing_ATT', lambda x: np.average(x, weights = stats1.loc[x.index, 'college_season_played'].fillna(0))),
    passing_COMPLETIONS = ('passing_COMPLETIONS', lambda x: np.average(x, weights = stats1.loc[x.index, 'college_season_played'].fillna(0))),
    passing_INT = ('passing_INT', lambda x: np.average(x, weights = stats1.loc[x.index, 'college_season_played'].fillna(0))),
    passing_PCT = ('passing_PCT', lambda x: np.average(x, weights = stats1.loc[x.index, 'college_season_played'].fillna(0))),
    passing_TD = ('passing_TD', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    passing_YDS = ('passing_YDS', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    passing_YPA = ('passing_YPA', lambda x: np.average(x, weights = stats1.loc[x.index, 'college_season_played'].fillna(0))),
    receiving_LONG = ('receiving_LONG', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    receiving_REC = ('receiving_REC', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    receiving_TD = ('receiving_TD', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    receiving_YDS = ('receiving_YDS', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    receiving_YPR = ('receiving_YPR', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    rushing_ATT = ('rushing_CAR', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    rushing_LONG = ('rushing_LONG', lambda x: np.average(x, weights = stats1.loc[x.index, 'college_season_played'].fillna(0))),
    rushing_TD = ('rushing_TD', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    rushing_YDS = ('rushing_YDS', lambda x: np.average(x, weights=stats1.loc[x.index, 'college_season_played'].fillna(0))),
    rushing_YPC = ('rushing_YPC', lambda x: np.average(x, weights = stats1.loc[x.index, 'college_season_played'].fillna(0)))
).reset_index(drop = False)

stats1
```
