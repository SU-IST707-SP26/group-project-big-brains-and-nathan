```python
import pandas as pd
combine = pd.read_csv("draft_combine_nflreadpy.csv")
combine.head()
```

```python
draft = pd.read_csv("draft_picks_df_cfbd.csv")
draft['position'] = draft['position'].replace('Quarterback','QB')
draft['position'] = draft['position'].replace('Running Back','RB')
draft['position'] = draft['position'].replace('Wide Receiver','WR')
draft['position'] = draft['position'].replace('Tight End','TE')
positions = ['QB','RB','WR','TE']
draft = draft[draft['position'].isin(positions)]
draft.head()
```

```python
stats = pd.read_csv("stats_merged_df.csv")
stats = stats[stats['position'].isin(positions)]
stats.head()
```

```python
merge1 = stats.merge(
    draft,
    how = "left",
    left_on = ["player_id"],
    right_on= ["college_athlete_id"]
)
merge1.head()
```

```python
final = merge1.merge(
    combine,
    how = "left",
    left_on= ["year","overall"],
    right_on= ["season","pick"]
)
final.head()
```

```python
final.to_csv("merged.csv")
```
