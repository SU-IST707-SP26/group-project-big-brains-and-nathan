```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('merged.csv')
data
```

```python
data.describe()
```

```python
positions = data['position_x'].value_counts()
positions.plot(kind = 'barh')
plt.title('Distribution of Positions')
plt.xlabel('Number of Players')
plt.ylabel('Position')
plt.show()
```

```python
positions = data['conference'].value_counts()
positions.plot(kind = 'barh')
plt.title('Number of Players by Conference')
plt.xlabel('Number of Players')
plt.ylabel('Conference')
plt.show()
```

```python
#top 15 teams by player count
positions = data['team_x'].value_counts().head(15)
positions.plot(kind = 'barh')
plt.title('Top 15 Teams by Amount of Players')
plt.xlabel('Number of Players')
plt.ylabel('Conference')
plt.show()
```

```python
#season distribution
plt.hist(data['height'])
plt.title('Age Distribution of Players')
plt.xlabel('Height (Inches)')
plt.ylabel('Number of Players')
plt.show()
```

```python
#age distribution - lot of NAs due to being NFL data
plt.hist(data['age'])
plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.show()
```

```python
#usage rate distribution 
plt.hist(data['usage_overall'])
plt.title('Usage Distribution of Players')
plt.xlabel('Usage Rate')
plt.ylabel('Number of Players')
plt.show()
```

```python
#total ppa distribution 
plt.hist(data['total_ppa_all'])
plt.title('PPA Distribution of Players')
plt.xlabel('Total PPA')
plt.ylabel('Number of Players')
plt.show()
```

```python
#Distribution of Passing TDs by QB
qbs = data[data['position_x'] == 'QB']

plt.hist(qbs['passing_TD'])
plt.title('Passing TD Distribution')
plt.xlabel('Total Passing TDs')
plt.ylabel('Number of Players')
plt.show()
```

```python
#QBs passing TD vs rushing TD

plt.scatter(qbs['passing_TD'], qbs['rushing_TD'])
plt.title('Quarterbacks: Passing TDs vs. Rushing TDs')
plt.xlabel('Total Passing TDs')
plt.ylabel('Total Rushing TDs')
plt.show()
```
