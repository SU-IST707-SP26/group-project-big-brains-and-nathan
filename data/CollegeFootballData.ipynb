```python
import pandas as pd
import time
import os
import cfbd
from cfbd.models.player_season_predicted_points_added import PlayerSeasonPredictedPointsAdded
from cfbd.rest import ApiException
from cfbd.models.season_type import SeasonType
from pprint import pprint

def load_cfbd_data():
  configuration = cfbd.Configuration(host = "https://api.collegefootballdata.com",
                                   #access_token = 'U20YXd7+EOCx8NNv8d2t2NIsMsImSIMxkBUeGaFvF+arxerB8O6XFUaG8pQDhgxv',
                                  access_token = 'UYYEE9HUhl8NhRa7McKXIW1CVe56Bn05WkHzQUh1paMwqn6ykJ7JO7rZvIbYQVnh')

  api_client = cfbd.ApiClient(configuration)

  ppa_instance = cfbd.MetricsApi(api_client)

  player_usage_instance = cfbd.PlayersApi(api_client)

  draft_instance = cfbd.DraftApi(api_client)

  stats_instance = cfbd.StatsApi(api_client)

  rankings_instance = cfbd.RankingsApi(api_client)

  ppa_records = []

  player_usage_list = []

  draft_picks_list = []

  player_stats_list = []

  rankings_list = []

  for year in range(2014,2025+1):
    ppa_response = ppa_instance.get_predicted_points_added_by_player_season(year = year)

    player_usage = player_usage_instance.get_player_usage(year = year)

    draft_picks = draft_instance.get_draft_picks(year = year)

    player_season_stats = stats_instance.get_player_season_stats(year = year)


    if year in [2014, 2019, 2020, 2024, 2025]:
      # final week of the regular season
      rankings = rankings_instance.get_rankings(year = year, week = 16,
                                                   season_type = SeasonType('regular'))
    else:
      rankings = rankings_instance.get_rankings(year = year, week = 15,
                                                   season_type = SeasonType('regular'))

    for player in ppa_response:
      ppa_records.append({
        'season': player.season,
        'player_id': player.id,
        'name': player.name,
        'position': player.position,
        'team': player.team,
        'conference': player.conference,
        'avg_ppa_passing_downs': player.average_ppa.passing_downs,
        'avg_ppa_standard_downs': player.average_ppa.standard_downs,
        'avg_ppa_third_down': player.average_ppa.third_down,
        'avg_ppa_second_down': player.average_ppa.second_down,
        'avg_ppa_first_down': player.average_ppa.first_down,
        'avg_ppa_rush': player.average_ppa.rush,
        'avg_ppa_var_pass': player.average_ppa.var_pass,
        'avg_ppa_all': player.average_ppa.all,
        'total_ppa_passing_downs': player.total_ppa.passing_downs,
        'total_ppa_standard_downs': player.total_ppa.standard_downs,
        'total_ppa_third_down': player.total_ppa.third_down,
        'total_ppa_second_down': player.total_ppa.second_down,
        'total_ppa_first_down': player.total_ppa.first_down,
        'total_ppa_rush': player.total_ppa.rush,
        'total_ppa_var_pass': player.total_ppa.var_pass,
        'total_ppa_all': player.total_ppa.all,
    })
    for player in player_usage:
      player_usage_list.append({
      'season': player.season,
      'player_id': player.id,
      'name': player.name,
      'position': player.position,
      'team': player.team,
      'conference': player.conference,
      'usage_passing_downs': player.usage.passing_downs,
      'usage_standard_downs': player.usage.standard_downs,
      'usage_third_down': player.usage.third_down,
      'usage_second_down': player.usage.second_down,
      'usage_first_down': player.usage.first_down,
      'usage_rush': player.usage.rush,
      'usage_var_pass': player.usage.var_pass,
      'usage_overall': player.usage.overall
    })
    for player in draft_picks:
      draft_picks_list.append({
        'college_athlete_id' : player.college_athlete_id,
        'nfl_athlete_id' : player.nfl_athlete_id,
        'college_id' : player.college_id,
        'college_team': player.college_team,
        'college_conference': player.college_conference,
        'nfl_team_id' : player.nfl_team_id,
        'nfl_team' : player.nfl_team,
        'position' : player.position,
        'name' : player.name,
        'position' : player.position,
        'year' : player.year,
        'round' : player.round,
        'pick_number' : player.pick,
        'overall' : player.overall,
        'height_in':player.height,
        'weight_lbs':player.weight,
        'pre_draft_position_ranking': player.pre_draft_position_ranking,
        'pre_draft_grade': player.pre_draft_grade,
      })
    for player in player_season_stats:
      player_stats_list.append(player.to_dict())

    for poll_week in rankings:
      for poll in poll_week.polls:
        if poll.poll == 'AP Top 25': # Filter for AP Top 25 Poll
          for rank_data in poll.ranks:
            rankings_list.append({
            'season': poll_week.season,
            'week':poll_week.week,
            'rank': rank_data.rank,
            'team_id' : rank_data.team_id,
            'school': rank_data.school,
            'points':rank_data.points,
            'first_place_votes':rank_data.first_place_votes
        })


  # Creating Predicted Points Added Dataframe
  ppa_df = pd.DataFrame(ppa_records)

  # Creating Player Usage Dataframe
  player_usage_df = pd.DataFrame(player_usage_list)

  # Creating Draft Picks Dataframe
  draft_picks_df = pd.DataFrame(draft_picks_list)

  # Creating player stats dataframe
  player_stats_df = pd.DataFrame(player_stats_list)

  player_stats_df = player_stats_df.rename(columns={'playerId' : 'player_id', 'player' : 'player_name', 'statType' : 'stat_type', 'stat' : 'stat_value'})

  # filtering for offensive plays
  player_stats_df = player_stats_df[player_stats_df.category.isin(['passing', 'receiving', 'rushing', 'fumbles'])]

  # filtering for WR, RB, QBs, & TE
  player_stats_df = player_stats_df[player_stats_df.position.isin(['WR', 'QB', "RB", 'TE'])]

  # Create a combined stat type to avoid duplicates when pivoting to wide format
  player_stats_df['combined_stat_type'] = player_stats_df['category'] + '_' + player_stats_df['stat_type']

  pivoted_player_stats = player_stats_df.pivot_table(
    index=['season', 'player_id', 'player_name', 'position', 'team', 'conference'],
    columns='combined_stat_type',
    values='stat_value',
    aggfunc='first' # takes the first value if there are duplicates
  ).reset_index()

  # Creating final rankings dataframe
  rankings_df = pd.DataFrame(rankings_list)

  return [ppa_df, player_usage_df, draft_picks_df, pivoted_player_stats, rankings_df]
```
